# Introduction to Flowcharts

Flowcharts are visual representations of a process, showing the sequence of steps needed to complete a task or solve a problem. They are widely used in various fields for planning, documenting, analyzing, and managing processes. Understanding the common symbols in flowcharts is crucial for interpreting and creating them. Here's a detailed explanation of some of the most commonly used flowchart symbols:







## 1. Rectangles (Process Steps)
- **Shape:** A rectangle or a box.
- **Purpose:** Represents a process step or an action to be taken.
- **Usage:** Each rectangle in a flowchart contains a brief description of the step it represents. For example, "Calculate Interest" or "Input Data".
- **Example in Flowchart:** A rectangle with the label "Review Application" indicates that the process involves reviewing an application at that step.

 ## 2. Diamonds (Decision Points)
- **Shape:** A diamond.
- **Purpose:** Used to represent a decision point in the process.
- **Usage:** The diamond shape contains a question or a decision to be made. The flowchart branches off from the diamond into different paths, each representing a possible answer or outcome of the decision.
- **Example in Flowchart:** A diamond labeled "Is the applicant eligible?" might have two arrows coming out of it, one labeled "Yes" leading to one process step and another labeled "No" leading to a different step or the end of the process.


## 3. Arrows (Flow Direction)
- **Shape:** A line with an arrowhead.
- **Purpose:** Indicates the direction of the flow of the process.
- **Usage:** Arrows connect the various symbols (like rectangles and diamonds) and show the sequence of steps and the flow from one step to another.
- **Example in Flowchart:** An arrow leading from a rectangle labeled "Submit Form" to a diamond labeled "Form Complete?" shows that after submitting the form, the next step is to check if the form is complete.

## **Other Common Symbols**

- **Ovals (Start/End):** Used to represent the start and end points of a flowchart. An oval contains words like "Start" or "End".
- **Parallelograms (Input/Output):** Used for steps involving input or output. For example, receiving data from a user or displaying results.
- **Circles (Connector):** Used to connect different parts of a flowchart, especially in complex diagrams where a process may continue on a different page or section.

**Tips for Using Flowchart Symbols**

- **Clarity:** Keep the labels inside the symbols clear and concise.
- **Consistency:** Use the symbols consistently throughout the flowchart.
- **Sequence:** Ensure that the arrows correctly represent the intended sequence of steps.

Flowcharts are powerful tools for simplifying complex processes and making them understandable at a glance. Knowing these symbols and how to use them effectively is key to creating useful and accurate flowcharts.

#Translating a flowchart into Python code
It involves converting the logical steps and decisions represented in the flowchart into Python syntax. Let's take the example of the flowchart for identifying prime numbers and translate it into Python code.


#### Flowchart Steps


1. **Start**
2. **Input a Number**
3. **Initialize Variables (Divisor = 2, IsPrime = True)**
4. **Check if Divisor <= (Number / 2)**
5. **If Yes, Check if Number % Divisor == 0**
6. **If Number % Divisor == 0, Set IsPrime = False**
7. **Increment Divisor by 1**
8. **Repeat Steps 4-7 until Divisor > (Number / 2)**
9. **Output Result (Prime or Not Prime)**
10. **End**


#### Python Translation

````python
# Step 1: Start
# Step 2: Input a Number
number = int(input("Enter a number: "))

# Step 3: Initialize Variables
divisor = 2
is_prime = True

# Step 4: Check if Divisor <= (Number / 2)
while divisor <= number / 2:
    # Step 5: Check if Number % Divisor == 0
    if number % divisor == 0:
        # Step 6: Set IsPrime = False
        is_prime = False
        break
    # Step 7: Increment Divisor by 1
    divisor += 1

# Step 9: Output Result
if is_prime and number > 1:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")

# Step 10: End
````

- **If Statement:** Inside the loop, it checks if the `number` is divisible by the `divisor`. If it is, `is_prime` is set to `False`, and the loop is exited.
- **Output:** After the loop, the code checks the value of `is_prime` to determine if the number is prime and prints the result.

## **Calculating Factorial**



- **Problem:** Calculate the factorial of a number (n! = n × (n-1) × (n-2) × ... × 1).

- **Flowchart Steps:**
 1. Start.
 2. Input the number (n).
 3. Initialize the result (factorial) to 1.
 4. Check if the number is greater than 1.
 5. Multiply the result by the number.
 6. Decrease the number by 1.
 7. Repeat steps 4-6 until the number is 1 or less.
 8. Output the factorial.
 9. End.


## **Finding the Least Common Multiple (LCM)**



- **Problem**: Find the LCM of two numbers.
- **Flowchart Steps:**
 1. Start.
 2. Input two numbers (a and b).
 3. Find the maximum of a and b.
 4. Check if the maximum is divisible by both a and b.
 5. If yes, output the maximum as LCM.
 6. If no, increase the maximum by 1 and repeat step 4.
End.

# **Introduction to Pseudocode**

Pseudocode in data science or web development is a technique used to describe the distinct steps of an algorithm in a way that’s easy for anyone with basic programming knowledge to understand. We use pseudocode in various fields of programming, whether it be app development, data science or web development. Pseudocode is a technique used to describe the distinct steps of an algorithm in a manner that’s easy to understand for anyone with basic programming knowledge. Although pseudocode is a syntax-free description of an algorithm, it must provide a full description of the algorithm’s logic so that moving from pseudocode to implementation is merely a task of translating each line into c

**Pseudocode Keywords**

SEQUENCE represents linear tasks sequentially performed one after the other.

WHILE a loop with a condition at its beginning.

REPEAT-UNTIL a loop with a condition at the bottom.

FOR another way of looping.

IF-THEN-ELSE a conditional statement changing the flow of the algorithm.

CASE the generalization form of IF-THEN-ELSE.




**Pseudocode Writing Rules**




*   Always capitalize the initial word (often one of the main six constructs).

*   Make only one statement per line.



*   Indent to show hierarchy, improve readability, and show nested constructs.

*   Always end multi-line sections using any of the END keywords (ENDIF, ENDWHILE, etc.).


* Use the naming domain of the problem, not that of the implementation. For instance: “Append the last name to the first name” instead of “name = first+ last.”

*   Keep it simple, concise and readable.
