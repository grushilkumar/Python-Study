## Programming Language

- Computer language that is used by programmers (developers) to communicate with computers.

-Set of instructions written in any specific language ( C, C++, Java, Python) to perform a specific task.

### Types of Programming Language : **Machine vs Assembly Language**

- **Machine Language:** Comprises binary code specific to a particular computer architecture. Uses sequences of 0s and 1s to represent operations and operands.

- **Assembly Language:** Slightly more human-readable than machine language, assembly language uses mnemonic codes for instructions.



assemblymachine.png

### Types of Programming Languages: **Compiled vs. Interpreted Languages**

- **Compiled Languages:** These languages, like C and C++, are translated entirely before execution. They go through a compilation process, converting the source code into machine code, resulting in faster execution as the entire code is pre-translated.

- **Interpreted Languages:** In contrast, interpreted languages such as Python and JavaScript are translated line-by-line during execution. They don't need a separate compilation step and offer flexibility but might be comparatively slower as the translation happens during runtime.

## Flow Control in Python

- Flow control in Python refers to the sequence in which statements are executed or controlled within a program.

- It allows programmers to dictate the order of execution based on certain conditions or loops.

Python offers three fundamental types of flow control:
- **Conditional Statements:** Control the flow of execution based on specific conditions

- **Transfer Statements** -  Alter the flow of execution by transferring control from one part of the program to another

- **Iterative Statements** - Allow the execution of a block of code repeatedly.

### Iterative Statements

- Iterative statements, or loops, allow the execution of a block of code repeatedly.

- Python offers for and while loops for iteration control.

#### for Loop

- **for loop:** Executes a block of code for a specified number of times or over elements of a sequence.

image.png

````python
# Iterating through a list of numbers
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num)
````

````python
# Iterating through characters in a string
message = "Hello, Python!"
for char in message:
    if char != ' ':  # Skip spaces
        print(char)
````

````python
# Using range() to generate a sequence of numbers
for i in range(1, 6):  # Range from 1 to 5 (6 is exclusive)
    print(i)
````

#### while Loop

- **while loop:** Executes a block of code as long as a specified condition is true.

image.png

````python
# Countdown from 5 to 1 using a while loop
count = 5
while count > 0:
    print(count)
    count -= 1
````

````python
# Asking for user input until a valid number is provided
valid_input = False
while not valid_input:
    user_input = input("Enter a number: ")
    if user_input.isdigit():  # Check if the input is a number
        print("Valid input!")
        valid_input = True
    else:
        print("Invalid input. Please enter a number.")
````

````python
# Calculate the sum of numbers from 1 to 5 using a while loop
total = 0
num = 1
while num <= 5:
    total += num
    num += 1
print("Sum of numbers from 1 to 5:", total)
````

#### Continue Statement

- Skips the rest of the loop code for the current iteration and moves to the next iteration.

**Usage:** The `continue` statement is employed within a loop, typically within an `if` statement that checks a specific condition.

**Purpose:** When the `continue` statement is executed, the current iteration is prematurely terminated, and the loop immediately proceeds to the next iteration.

**Example:**
```python
for i in range(5):
    if i == 3:
        continue
```

In this example, when `i` equals 3, the current iteration is skipped, and the loop continues with the next value of `i` (4 and 5). The `continue` statement is a useful tool for controlling the flow of loops and skipping specific iterations based on a condition.

Question:
- Write a program that continuously asks the user to provide an input number.
- Print the square of that input number.
- The program should stop if the user enters 5.
- In case user enters any multiple of 4, skip it.

````python
while True:
    x = int(input())

    if x == 5:
        break

    if x % 4 == 0:
        continue

    print(x ** 2)
````

#### Pass Statement
pass in Python is a null statement which the interpreter simply 'moves over' while executing the code. It can be used in a variety of scenarios where you want to leave parts of your code undefined (such as if-else, loops, classes or methods) but the Python syntax doesn't allow it.

````python
x = 10

if x % 4 == 0:
    pass
````

````python
for i in range(10):
    pass
````

#### Break Statement

Terminates the loop it is contained in, allowing the program to exit the loop prematurely.


**Usage:** The `break` statement is utilized within a loop, often accompanied by a conditional statement (such as an `if` statement) that checks a specific condition.

**Purpose:** When the `break` statement is executed, the loop is immediately terminated, and the program proceeds with the next statement after the loop.

**Example:**
```python
for i in range(5):
    if i == 3:
        break
```

In this example, the loop stops when `i` equals 3, and the remaining numbers (4 and 5) are not iterated over. The `break` statement is a valuable tool for controlling the flow of loops and prematurely exiting a loop when a certain condition is met.

````python
for i in range(10):
    print(i)

    if i == 5:
        break

    print("After the if")

print("\nOutside the loop")
````

#### Nested Loops

**Usage:** Nested loops involve placing one loop inside another, providing a powerful way to iterate through complex data structures, such as multi-dimensional lists or matrices.

**Purpose:** Nested loops are commonly used for tasks like traversing rows and columns in a 2D array or searching for elements in a grid.

**Example:**
```python
for i in range(3):
    for j in range(2):
        print(i, j)
```
In this example, two nested `for` loops are used. The outer loop iterates three times (i = 0, 1, 2), and for each iteration of the outer loop, the inner loop iterates twice (j = 0, 1). This results in a total of six combinations of `i` and `j` being printed.

````python
for j in range(5):
    print("*", end='')
````

````python
for j in range(5):
    print("*", end='')

print()

for j in range(5):
    print("*", end='')

print()

for j in range(5):
    print("*", end='')
````

````python
for i in range(3):
    for j in range(5):
        print("*", end='')
    print()
````

##### Nested Loops: Stair Pattern

````python
N = 5

for i in range(1, N+1):
    for j in range(i):
        print("*", end='')
    print()
````

##### Nested Loop: Reverse Stair Pattern

````python
N = 15

for i in range(1, N+1):
    # For each row, print (N-i+1) cols
    for j in range(N-i+1):
        print('*', end='')
    print()
````

## GCD Calculation
Given two integer inputs, write a program to calculate GCD of these two numbers.


````python
A = 200
B = 80

# Smallest = 1
# Largest = min(a, b)
# I am looking for the highest common factor

for i in range(min(A, B), 0, -1):
    # We wish to find a common factor of A & B
    # Check if (i is a factor of A) and (i is a factor of B)
    if (A % i == 0) and (B % i == 0):
        gcd = i
        break

print(gcd)
````
